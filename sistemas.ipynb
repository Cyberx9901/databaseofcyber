```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
dados = pd.read_csv('https://raw.githubusercontent.com/Cyberx9901/databaseofcyber/main/dados_distros.csv', sep=';')
```

# ***As 30 Distribuições GNU/Linux mais pesquisadas*** #

```python
dados.head(30)
```

```python
fig = plt.figure(figsize = (25,5))
plt.bar(dados.groupby(by='AMBIENTE DESKTOP(PADRÃO)').count().index, dados.groupby(['AMBIENTE DESKTOP(PADRÃO)']).size())
plt.tick_params(axis='x', labelsize=17)
plt.ylabel('Quantidade por distro', size='14')
plt.title('Ambientes de Trabalho', size='18')
plt.show()
```

```python
plt.figure(figsize=(5, 5))

dados['BASE'].value_counts()[:30].plot(kind='pie', 
                                            autopct='%2.1f%%',                                            
                                            startangle=0,
                                            pctdistance=0.8,
                                            fontsize=12)
centre_circle = plt.Circle((0,0), 0.30,fc='white')
fig = plt.gcf()
fig.gca().add_artist(centre_circle)

plt.title('Distribuições utilizadas como base')
plt.tight_layout()
plt.axis('off') 
plt.show()
```

```python

plt.bar(dados.groupby(by='ARQUITETURA').count().index, dados.groupby(['ARQUITETURA']).size())

plt.xlabel('Arquiteturas', size='14')
plt.ylabel('Quantidade de suporte por Distros', size='14')
plt.title('Arquiteturas por Distro', size='18')
plt.show()
```

```python
plt.figure(figsize=(10, 10))

dados['ORIGEM'].value_counts()[:30].plot(kind='pie', 
                                            autopct='%2.1f%%',                                            
                                            startangle=0,
                                            pctdistance=0.8,
                                            fontsize=12)
centre_circle = plt.Circle((0,0), 0.30,fc='white')
fig = plt.gcf()
fig.gca().add_artist(centre_circle)

plt.title('Origem do desenvolvimento da Distro')
plt.tight_layout()
plt.axis('off') 
plt.show()
```

Podemos concluir com os gráficos que a maior parte está dividida entre distros com ambientes de trabalho padronizados, porém a escolha mais recorrente para as distros é não ter um único ambiente de trabalho padrão.

A respeito da base da distro, a maioria é baseada em algum outro sistema, tendo a maioria como alvo o Ubuntu, porém, se compararmos individualmente há mais distros independentes do que de qualquer outra distribuição, entre as mais procuradas.

Distribuições que contenham suporte a multiarquitetura também são a grande maioria.

O pais aonde mais se origina distros é nos Estados Unidos, seguido da Irlanda, porém a grande maioria das distribuições vem de fora desses paises.

```python
obj='AMBIENTE DESKTOP(PADRÃO)'
acessos_diarios = dados.loc[:, [obj, 'ACESSOS DIARIOS']].groupby(by=obj).sum().loc[:, 'ACESSOS DIARIOS']

fig = plt.figure(figsize = (25,5))
plt.tick_params(axis='x', labelsize=15)
plt.bar(dados.groupby(by=obj).count().index, acessos_diarios)
plt.xlabel('DE', size='17')
plt.ylabel('Quantidade de acessos diários', size='14')
plt.title('Popularidade de Ambiente Desktop Padrão', size='18')
plt.show()
```

```python
obj='BASE'
acessos_diarios = dados.loc[:, [obj, 'ACESSOS DIARIOS']].groupby(by=obj).sum().loc[:, 'ACESSOS DIARIOS']

plt.bar(dados.groupby(by=obj).count().index, acessos_diarios)
plt.xlabel('Distros', size='14')
plt.ylabel('Quantidade de acessos diários', size='14')
plt.title('Popularidade por Distro', size='18')
plt.show()
```

```python
obj='ARQUITETURA'
acessos_diarios = dados.loc[:, [obj, 'ACESSOS DIARIOS']].groupby(by=obj).sum().loc[:, 'ACESSOS DIARIOS']

plt.bar(dados.groupby(by=obj).count().index, acessos_diarios)
plt.xlabel('Arquiteturas', size='14')
plt.ylabel('Quantidade de acessos diários', size='14')
plt.title('Popularidade por arquiteturas', size='18')
plt.show()
```

```python
obj='ORIGEM'
acessos_diarios = dados.loc[:, [obj, 'ACESSOS DIARIOS']].groupby(by=obj).sum().loc[:, 'ACESSOS DIARIOS']

fig = plt.figure(figsize = (40,10))
plt.tick_params(axis='x', labelsize=15)
plt.bar(dados.groupby(by=obj).count().index, acessos_diarios)
plt.xlabel('Paises', size='17')
plt.ylabel('Quantidade de acessos diários', size='17')
plt.title('Popularidade por pais', size='18')
plt.show()
```

Um detalhe a ser notado é que por mais que existam mais distros sem ambientes de trabalho padrão entre as 30 distribuições mais populares, a verdade é que o que mais se repete o interesse sobre distros com XFCE, como pode ser verificado de acordo com os gráficos.

O mesmo acontece com a base dos sistemas, por mais que existam mais sistemas com bases independentes, os usuários parecem ter mais interesse em  distros com base Ubuntu, embora casos como suporte a multiarquitetura mantenham uma semelhança entre a popularidade, e a quantidade de distribuições que as suportam, assim como a popularidade de sistemas desenvolvidos nos Estados Unidos, e a quantidade produzida.
